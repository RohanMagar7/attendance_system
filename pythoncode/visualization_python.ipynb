

##### **colored scatterplots**

```python
iris = sns.load_dataset('iris')
iris
iris.sample(5)
```

```python
# replacing 
iris['species'] = iris['species'].replace({'versicolor' : 1 ,
                                           'setosa' : 0 , 
                                           'virginica' : 2
                                           })

iris.sample(5)
```

```python
plt.scatter(iris['sepal_length'] , iris['petal_length'] , 
            c = iris['species'])

plt.xlabel('sepal length')
plt.ylabel('petal length')

plt.show()
```

```python
plt.scatter(iris['sepal_length'] , iris['petal_length'] , 
            c = iris['species'] , cmap='winter')

plt.xlabel('Sepal length')
plt.ylabel('petal length') 

plt.show()
```

```python
plt.scatter(iris['sepal_length'] , iris['petal_length'] , 
            c = iris['species'] , cmap = 'viridis')

plt.xlabel('sepal length')
plt.ylabel('petal length')

plt.colorbar()

plt.show()
```

##### **alpha**

```python
plt.scatter(iris['sepal_length'] , iris['petal_length'] , 
            c = iris['species'] , cmap = 'jet' , 
            alpha = 0.5)

plt.xlabel('sepal length')
plt.ylabel('petal length')

plt.colorbar()

plt.show()
```

##### **plot size**

```python
plt.figure(figsize = (15 , 7 ))
plt.scatter(iris['sepal_length'] , iris['petal_length'] , 
            c = iris['species'] , cmap = 'jet' , alpha= 0.5)

plt.xlabel('sepal lenght')
plt.ylabel('petal length')

plt.colorbar()

plt.show()
```

```python
root_dir = 'F:/attendance system/pythoncode/'

batter = pd.read_csv(root_dir + 'batter.txt')
batter
```

```python
sample_df = batter.sample(25,random_state = 29)
sample_df.shape
sample_df
```

```python
plt.scatter(sample_df['avg'] , sample_df['strike_rate'] , color = 'green')
```

```python
plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'] , sample_df['strike_rate'] , 
            s= sample_df['runs'] )

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i] ,
             sample_df['strike_rate'].values[i] , 
             sample_df['batter'].values[i])
```

##### Horizontal and vertical Line

```python
plt.figure(figsize = (18 , 10 ))

plt.axhline(130 , color = 'red' , linewidth = 5)

plt.scatter(sample_df['avg'] , sample_df['strike_rate'] , 
            s = sample_df['runs'] , color = 'gold')


for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i] , 
             sample_df['strike_rate'].values[i],
             sample_df['batter'].values[i])
```

```python
plt.figure(figsize = (18 , 10 ))

plt.axvline(30 , color = 'red' , linewidth = 10)

plt.scatter(sample_df['avg'] , sample_df['strike_rate'] , 
            s = sample_df['runs'] , color = 'gold')


for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i] , 
             sample_df['strike_rate'].values[i],
             sample_df['batter'].values[i])
```

```python
plt.figure(figsize = (18 , 10 ))

plt.axhline(130 , color = 'red' , linewidth = 5)
plt.axvline(40 , color = 'red' , linewidth = 6)

plt.scatter(sample_df['avg'] , sample_df['strike_rate'] , 
            s = sample_df['runs'] , color = 'gold')


for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i] , 
             sample_df['strike_rate'].values[i],
             sample_df['batter'].values[i])
```

```python
plt.figure(figsize = (18 , 10 ))

plt.axhline(130 , color = "red" , linewidth = 5)
plt.axhline(140 , color = 'green' , linewidth = 5)
plt.axvline(35 , color = 'red' , linewidth = 5)


plt.scatter(sample_df['avg'] , sample_df['strike_rate'] , 
            s = sample_df['runs'] , color= 'pink')


for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],
             sample_df['strike_rate'].values[i], 
             sample_df['batter'].values[i])
```
