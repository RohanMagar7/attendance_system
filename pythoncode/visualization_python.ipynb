```python
# from a padans dataframe 
batsman = pd.read_csv('F:/attendance system/pythoncode/svskohli.csv')
batsman
```

```python
plt.plot(batsman['Year'] , batsman['V Kohli'])
```

```python
plt.plot(batsman['Year'] , batsman['V Kohli'])
# plt.plot(batsman['year'] , batsman['RG Sharma'])
plt.plot(batsman['Year'] , batsman['RG Sharma'])
```

```python
# labels Title 
plt.plot(batsman['Year'] , batsman['V Kohli'])
plt.plot(batsman['Year'] , batsman['RG Sharma']) 

plt.title('Rohit Sharma VS virat kohli Career difference')
plt.xlabel('season')
plt.ylabel('Runs scored')
```

```python
# colors ( hex ) # Hex colors
plt.plot(batsman['Year'] , batsman['V Kohli'] , color = '#ADAA19')
plt.plot(batsman['Year'] , batsman['RG Sharma'] , color = '#FC00D6')

plt.title('Rohit Sharma VS virat kohli Career difference')
plt.xlabel('season')
plt.ylabel('Runs scored')
 
```

```python
plt.plot(batsman['Year'] , batsman['V Kohli'] , color = '#ADAA19' , linestyle = 'dashdot')
plt.plot(batsman['Year'] , batsman['RG Sharma'] , color = '#FC00D6' )

plt.title('Rohit Sharma VS virat kohli Career difference')
plt.xlabel('season')
plt.ylabel('Runs scored')
```

#### **dotted lines**

```python
plt.plot(batsman['Year'] , batsman['V Kohli'] , color = '#ADAA19' , linestyle = 'dotted')
plt.plot(batsman['Year'] , batsman['RG Sharma'] , color = '#FC00D6' , linestyle = 'dotted')

plt.title('Rohit Sharma VS virat kohli Career difference')
plt.xlabel('season')
plt.ylabel('Runs scored')
```

##### **LIneWidth**

```python
plt.plot(batsman['Year'] , batsman['V Kohli'] , color = '#ADAA19' , linestyle = 'solid' , linewidth = 3)
plt.plot(batsman['Year'] , batsman['RG Sharma'] , color = '#FC00D6', linestyle = 'dashdot' , linewidth = 2)

plt.title('Rohit Sharma VS virat kohli Career difference')
plt.xlabel('season')
plt.ylabel('Runs scored')
```

##### **Marker ( Size )**

```python
plt.plot(batsman['Year'] , batsman['V Kohli'] , color = '#ADAA19' , linestyle = 'solid' , linewidth = 3 , marker = '>', markersize = 10 )

plt.plot(batsman['Year'] , batsman['RG Sharma'] , color = '#FC00D6', linestyle = 'dashdot' , linewidth = 2 , marker = 'D')

plt.title('Rohit Sharma VS virat kohli Career difference')
plt.xlabel('season')
plt.ylabel('Runs scored')
```

```python
plt.plot(batsman['Year'] , batsman['V Kohli'] , color = '#ADAA19' , linestyle = 'solid' , linewidth = 3 , marker = '>', markersize = 10 , label = 'Virat Kohli ')

plt.plot(batsman['Year'] , batsman['RG Sharma'] , color = '#FC00D6', linestyle = 'dashdot' , linewidth = 2 , marker = 'D' , label = 'RG Sharma ')

plt.title('Rohit Sharma VS virat kohli Career difference')
plt.xlabel('season')
plt.ylabel('Runs scored')

plt.legend() # loc = best
```

```python
# legend location 
plt.plot(batsman['Year'] , batsman['V Kohli'] , color = '#ADAA19' , linestyle = 'solid' , linewidth = 3 , marker = '>', markersize = 10 , label = 'Virat Kohli ')

plt.plot(batsman['Year'] , batsman['RG Sharma'] , color = '#FC00D6', linestyle = 'dashdot' , linewidth = 2 , marker = 'D' , label = 'RG Sharma ')

plt.title('Rohit Sharma VS virat kohli Career difference')
plt.xlabel('season')
plt.ylabel('Runs scored')

plt.legend(loc = 'lower right') # loc = best
```
